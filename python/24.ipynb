```python
df['X'] = df['X'].astype('object')
df['Y'] = df['Y'].astype('object')
df['FFMC'] = df['FFMC'].astype('object')
df['DMC'] = df['DMC'].astype('object')
df['ISI'] = df['ISI'].astype('object')
df['DC'] = df['DC'].astype('object')
df['RH'] = df['RH'].astype('object')
df['wind'] = df['wind'].astype('object')
df['rain'] = df['rain'].astype('object')
# df['area'] = df['area'].astype('object')
df['temp'] = df['temp'].astype('object')
# df['area'] = df['area'].astype('float64')
```

```python
df
```

```python
df['area'].max()
```

```python
# df['area'] = pd.to_numeric(df['area'], errors='coerce')  # Convert 'area' column to numeric

for i in df['area'].values:
    if i == 0.0:
        df['area'] = df['area'].replace(i, 'not affected')
    elif (i > 0.0) and (i <= 250.0):
        df['area'] = df['area'].replace(i, 'partially affected')
    else:
        df['area'] = df['area'].replace(i, 'mostly affected')
```

```python
df
```

```python
subset1 = df[df['area'] == 'not affected']
subset1
```

```python
subset2 = df[df['area'] == 'partially affected']
subset2
```

```python
subset3 = df[df['area'] == 'mostly affected']
subset3
```

```python
merge = pd.concat([subset2, subset3], axis = 0)
merge
```

```python
sort = merge.sort_values(by = ['temp', 'wind', 'area'], ascending = True)
sort.head(20)
```
